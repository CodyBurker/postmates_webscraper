```python
import pandas as pd
```

```python
# Read in joined_wDistance.csv
df = pd.read_csv('joined_wDistance.csv')
# Drop first two columns
df = df.drop(df.columns[0:2], axis=1)
df
```

```python
# Readin ACS demographic data
acs = pd.read_csv('ACS/acs_demographic_data.csv')
# Drop first column
acs = acs.drop(acs.columns[0], axis=1)
acs
```

```python
# Rename GEOID to Geoid in acs
acs = acs.rename(columns={'GEOID': 'Geoid'})
# Left join df and acs on Geoid
df_merged = pd.merge(df, acs, on='Geoid', how='left')
df_merged
```

```python
# Read 'ACS/holc_grade.csv' using correct encoding
holc_grade = pd.read_csv('ACS/holc_grade_utf8.csv')
# holc = pd.read_csv('ACS/holc_grade.csv', sep=',')
holc_grade.head()
```

```python
# Read ACS/holc_tract_lookup_utf8.csv
holc_tract = pd.read_csv('ACS/holc_tract_lookup_utf8.csv')
holc_tract.head()
```

```python
# For each row in df_merged, get all the values in the column 'Holc Grade' where Geoid is equal to the Geoid in the row
df_merged['Holc Grade'] = df_merged.apply(lambda row: holc_grade[holc_grade['Geoid'] == row['Geoid']]['Holc Grade'].values, axis=1)
```
