```python
import pandas as pd
import os
```

```python
# Folder to read all files from
raw_folder = "11-19-2021 5pm ET"
# Get list of file names in folder
file_list = os.listdir(raw_folder)
# Loop over files in folder and combine to one dataframe
df = pd.DataFrame()
for file in file_list:
    # Read in file
    file_df = pd.read_csv(raw_folder+'/'+file)
    # Combine to one dataframe
    df = df.append(file_df)
df
```

```python
# Print percent where Wait Time is not NaN
print("Wait time complete:")
print(df['Wait Time'].count() / len(df['Wait Time']) * 100)
# Do the same for Name
print("Name Complete")
print(df['Name'].count() / len(df['Name']) * 100)
# Do the same for Delivery Fee
print("Delivery Fee Complete")
print(df['Delivery Fee'].count() / len(df['Delivery Fee']) * 100)
# Do the same for Link
print("Link Complete")
print(df['Link'].count() / len(df['Link']) * 100)
# Do the same for Delivery Address (sanity check)
print("Delivery Address Complete")
print(df['Delivery Address'].count() / len(df['Delivery Address']) * 100)
# Do the same for Geoid
print("Geoid Complete")
print(df['Geoid'].count() / len(df['Geoid']) * 100)
```

```python
# Read in OAK_Berk_geocodio_filtered.csv
oak_berk_geocodio_filtered = pd.read_csv('OAK_Berk_geocodio_filtered.csv')
oak_berk_geocodio_filtered
```

```python
# Left join df with oak_berk_geocodio_filtered on Geoid
df_merged = df.merge(oak_berk_geocodio_filtered, on='Geoid', how='left')
df_merged.columns
df_merged.to_csv(raw_folder + '/' + 'df_merged.csv')
```
