```python
import pandas as pd
import os
```

```python
# Folder to read all files from
raw_folder = "11-19-2021 11am ET"
# Get list of file names in folder
file_list = os.listdir(raw_folder)
# Loop over files in folder and combine to one dataframe
df = pd.DataFrame()
for file in file_list:
    # Read in file
    file_df = pd.read_csv(raw_folder+'/'+file)
    # Combine to one dataframe
    df = df.append(file_df)
df
```
